## Concatenating

1. Read `crew_data.tsv.gz` to DataFrame named `movie_crew_df` with columns `ID, director_ids` indexed by `ID` (forget the rest of the columns)
2. Read `person_data.tsv.gz` to DataFrame named `person_df` with columns `person_ID, name` indexed by `person_ID` (forget the rest of the columns)
3. Merge `movie_crew_df` into `df` with `ID`s and drop the columns without necessary data (hint `outer`-join)
4. Remove movies with more than one director and rename `director_ids` => `director_id`
5. Merge `df` with `person_df` on `director_id`, remove `director_id` and `person_ID` columns, rename `name` => `director`

Read crew_data.tsv.gz to DataFrame named movie_crew_df with columns ID, director_ids indexed by ID (forget the rest of the columns).

```python

movie_crew_df= pd.read_csv('crew_data.tsv', delimiter="\t")
movie_crew_df=movie_crew_df.rename(columns={'tconst':'ID','directors':'director_ids'})
movie_crew_df.set_index("ID")
movie_crew_df.head()
```

```python
movie_crew_df=movie_crew_df.drop(columns='writers')
movie_crew_df.head()
```

```python
import pandas as pd
person_df = pd.read_csv("name_data.tsv", delimiter= "\t", header = 0, usecols=(0,1), names=["person ID","name"])
person_df.head()
```

```python
df=pd.merge(movie_crew_df, df, on='ID')
df.head()
```

```python
indexNames = df[((df['director_ids'].str.len()) > 9)].index
df.drop(indexNames , inplace=True)
```

```python
df.head(20)
```

```python
df=df.rename(columns={'director_ids':'director_id'})
df.head()
```

```python
df.shape
```

```python
df2= pd.merge(df, person_df, left_on="director_id", right_on= "person ID", how= "inner")
df2
```
